# Visualize and Analyze error results

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import os
import zipfile
```

```python
test_path = "ORB_SLAM_TEST/"
files = os.listdir(test_path)

file_dirs = [name for name in files if '.csv' not in name and 'ipynb' not in name]
```

```python
file_dirs
```

```python
dataset_head = 'rgbd_dataset_freiburg1'
dataset_tail = 'xyz'

dataset_chosen_dir = test_path + dataset_head + "_" + dataset_tail +"/"
```

```python
error_files = os.listdir(dataset_chosen_dir)
```

```python
error_dirs = sorted([error_dir for error_dir in error_files if "." not in error_dir])
```

```python
error_gfs_dirs = [error_dir for error_dir in error_dirs if '_gfs' in error_dir]
error_nogfs_dirs = [error_dir for error_dir in error_dirs if '_nogfs' in error_dir]
```

```python
error_list = {'gfs':[],'nogfs':[]}
error_mean = {'gfs':[],'nogfs':[]}
error_std = {'gfs':[],'nogfs':[]}
plt.figure(figsize=(16,8))
for gfs_dir in error_dirs:
    dir_path_full = dataset_chosen_dir + gfs_dir + '/'
    error_new = np.load(dir_path_full + 'error_array.npy')
    plt.plot(error_new,label=gfs_dir)
    error_list['gfs'] += [error_new]
    
plt.legend()
plt.show()
```
